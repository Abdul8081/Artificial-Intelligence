```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread('images/sky.jpg')
img = cv.cvtColor(image, cv.COLOR_RGB2BGR)

plt.imshow(img)
plt.axis('off')
```

```python
#Filtering of the image, showing then saving]
img = np.array(img)
# print(img, np.dtype)

#Filtering to blue
blue_filtered = img * [0., 0., 1.]
plt.imshow(blue_filtered)
plt.title('Filtered Blue part of the Image')
plt.axis("off")


```

```python
# Saving of the Image
# cv.imwrite('images/blue_filtered.png', blue_filtered)  
read_image = cv.imread('images/blue_filtered.png')
actual_read_image = cv.cvtColor(read_image, cv.COLOR_BGR2RGB)

plt.imshow(actual_read_image)
plt.title("Reading of the Saved Image from directory")
plt.axis('off')
```

```python
# Rotation, Blurring and Denoisation of the image 
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

image = cv.imread('images/sky.jpg')
#converting image to the actual image
image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
plt.imshow(image)
plt.axis('off')
plt.title("Actual Image")
```

```python
#Rotation 
def rotate(image): 
    rotated_image = cv.rotate(image, cv.ROTATE_90_CLOCKWISE)
    cv.imwrite('images/rotated.png', rotated_image)

def blurred(image):
    blurred_image = cv.blur(image, (5 ,5))
    cv.imwrite('images/blured.png',blurred_image)

def denoised(image):
    median_blur = cv.medianBlur(image, 5)
    cv.imwrite('images/denoised.png',median_blur)

rotate(image)
# blurred(image)
# denoised(image)
```

# OME-PIFF Image 
OME-TIFF is a specializedc file formate designed for storing and sharing multidimensional scientific imaging data, particularly in biological and microscopy research. It builds on the TIFF standard by embedding rich metadata in the form of OME-XML within the image file headers. This allows it to store not just pixel data but also extensive information about th image acquisition process, such as microscope setting, dimentions. 

### It supports  
1. MULTIDIMENSIONAL : It can be used to Stores 5D+ data (X, Y, Z, time, channels, and more) in a single file or across multiple files.
2. It can be a single master file or split into muliple files with consistent metadata linking them via UUIDs.
