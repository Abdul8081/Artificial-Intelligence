```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv


img = cv.imread('images/sky.jpg') #this will not show the real image.
#print(img)
plt.imshow(img) 
plt.axis("off")
```

```python
actual_image = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(actual_image)
plt.axis("off")
```
